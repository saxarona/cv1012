# Integrals and numerical integration examples

```python
import numpy as np
import sympy as sym
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns

mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')
sns.set_style('darkgrid')
```

```python
def f(x):
    return 3 * x ** 2 + 7

def deriv(f, x, dx=0.000001):
    return(f(x + dx) - f(x)) / dx
```

```python
x = np.linspace(-10, 65)
y = f(x)
x_0 = 40
y_0 = f(x_0)

tangent = deriv(f, x_0) * (x - x_0) + y_0

fig, ax = plt.subplots(1,1)

sns.lineplot(x, y, color='red')
plt.plot(x, tangent, "b")
plt.title(r"$f(x) = 3x^2 + 7 $ with its tangent line at 40")
plt.savefig("tangent.pdf")
```

```python
plt.clf()
x_1 = -10
y_1 = f(x_1)
x_2 = 65
y_2 = f(x_2)

linex = np.array([x_1, x_2])
liney = np.array([y_1, y_2])

fig, ax = plt.subplots(1,1)

sns.lineplot(x, y, color='red')
plt.plot(linex, liney, '.-k', markersize=10)
plt.plot(x, tangent, "b")
plt.title(r"$f(x) = 3x^2 + 7 $ with difference using -10 and 65")
plt.savefig("diff01.pdf")
```

Example 21.2 in Chapra, Canale

```python
def exf(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
x = sym.symbols('x')
symexf = 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
symexf
```

The numerical approach

```python
fig, ax = plt.subplots(1,1)
mesh = np.linspace(-2, 2, 1000)
boxes = np.linspace(0, 0.8, 3)
sns.lineplot(mesh, exf(mesh), color='red')
plt.plot(boxes, exf(boxes), '.-k', markersize=10)
plt.fill_between(boxes, exf(boxes), color=(0.33, 0.33, 0.33, 0.2))
plt.ylim(-1, 5)
plt.suptitle(r"$f(x) = 400 x^{5} - 900 x^{4} + 675 x^{3} - 200 x^{2} + 25 x + 0.2$")
plt.savefig("two-segment.pdf")
```

```python
fig, ax = plt.subplots(1,1)
mesh = np.linspace(-2, 2, 1000)
boxes = np.linspace(0.001, 0.8, 3)
plt.plot(mesh, exf(mesh), color='red')
sns.regplot(boxes, exf(boxes), order=2, ci=None)
plt.fill_between(boxes, exf(boxes), color=(0.33, 0.33, 0.33, 0.2))
plt.ylim(-1, 5)
plt.suptitle(r"$f(x) = 400 x^{5} - 900 x^{4} + 675 x^{3} - 200 x^{2} + 25 x + 0.2$")
plt.savefig("simpsononethird.pdf")
```

```python
fig, ax = plt.subplots(1,1)
mesh = np.linspace(-2, 2, 1000)
boxes = np.linspace(0.001, 0.4, 3)
boxes1 = np.linspace(0.4, 0.8, 3)
plt.plot(mesh, exf(mesh), color='red')
sns.regplot(boxes, exf(boxes), order=2, ci=None)
sns.regplot(boxes1, exf(boxes1), order=2, ci=None)
plt.fill_between(boxes, exf(boxes), color=(0.33, 0.33, 0.33, 0.2))
plt.fill_between(boxes1, exf(boxes1), color=(0.33, 0.33, 0.33, 0.2))
plt.ylim(-1, 5)
plt.suptitle(r"$f(x) = 400 x^{5} - 900 x^{4} + 675 x^{3} - 200 x^{2} + 25 x + 0.2$")
plt.savefig("simpsonmultionethird.pdf")
```

The 'analytical' approach

```python
n = 2
a = 0
b = 0.8
midpoint = (b - a) / n
I = (b - a) * (exf(a) + 2 * exf(midpoint) + exf(b)) / (2 * n)
I
```

```python
anI = sym.integrate(symexf, (x, a, b))
anI
```

Calculating error by total error formula

```python
(anI - I)/anI * 100
```

Error by average second derivative

```python
exfbiprime = sym.diff(symexf, x, 2)
exfbiprime
```

we get the average second derivative by using...

```python
avgfbiprime = sym.integrate(exfbiprime, (x, a, b)) / (b - a)
avgfbiprime
```

Which we later include in our calculation for error:

```python
np.abs((b - a) ** 3 / (12 * n ** 2) * avgfbiprime)
```
